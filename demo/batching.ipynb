```python
from vipy.annotation import googlesearch
from vipy.image import ImageCategory, Batch
from vipy.visualize import montage
```

```python
urls = googlesearch('owls')
imlist = [ImageCategory(url=url, category='owl').load(ignoreErrors=True) for url in urls]
```

```python
imbatch = Batch([im for im in imlist if im.isloaded()])
imlist = imbatch.mindim(224).centersquare().rgb().list() 
```

```python
print(imbatch.torch().shape)
```

```python
montage(imlist, gridrows=128, gridcols=128, do_plot=True);
```
