```python
import vipy.video
import vipy.object
from vipy.geometry import BoundingBox
```

```python
mp4file = '../test/Video.mp4'
vid = vipy.video.Scene(filename=mp4file, tracks=[vipy.object.Track('RandomBox1', frames=[0,200], boxes=[BoundingBox(xmin=0,ymin=0,width=200,height=400), BoundingBox(xmin=100,ymin=200,width=400,height=100)]),
                                                 vipy.object.Track('RandomBox2', frames=[0,200], boxes=[BoundingBox(xmin=100,ymin=200,width=300,height=400), BoundingBox(xmin=400,ymin=300,width=100,height=400)])])
print(vid)
```
